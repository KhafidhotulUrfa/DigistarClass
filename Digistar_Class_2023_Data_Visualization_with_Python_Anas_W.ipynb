**IMPORT LIBRARIES**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

**BAR PLOT**

```python
# Sample data for the bar plot
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4'] # sumbu X
values = [25, 40, 30, 50] #sumbu Y

# Creating the bar plot
plt.bar(categories, values)

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Ini Barchart')

# Displaying the plot
plt.show()
```

```python
# Sample data for the count plot
data = ['Category A', 'Category B', 'Category A', 'Category C', 'Category B', 'Category A', 'Category A', 'Category B', 'Category C']

# Create the count plot
plt.figure(figsize=(8, 6))
sns.countplot(x=data)

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Count Plot Example')

# Displaying the plot
plt.show()
```

```python
# Sample data for the count plot
data = ['Category A', 'Category B', 'Category A', 'Category C', 'Category B', 'Category A', 'Category A', 'Category B', 'Category C']

# Create the horizontal count plot
plt.figure(figsize=(8, 6))
sns.countplot(y=data)

# Adding labels and title
plt.xlabel('Count')
plt.ylabel('Categories')
plt.title('Horizontal Count Plot Example')

# Displaying the plot
plt.show()
```

**LINE PLOT**

```python
# Sample data for the line plot
x = [1, 2, 3, 4, 5]
y = [10, 15, 20, 25, 30]

# Creating the line plot
plt.plot(x, y)

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot Example')

# Displaying the plot
plt.show()
```

**SCATTER PLOT**

```python
# Creating a scatter plot using Seaborn
tips = sns.load_dataset('tips') #using sample dataset from libraries
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex') # x = variable we put as X axis, y = variable we put as Y axis, data = dataset we use, hue = what variable we use to determine color
plt.xlabel('Total Bill') # xlabel = x axis name / label
plt.ylabel('Tip') # ylabel = y axis name / label
plt.title('Tip vs. Total Bill') # title = plot title / graph name

# Saving the plot with a name (e.g., "scatter_plot.png")
plt.savefig('scatter_plot.png') #savefig = function to save our plot to image format

# Displaying the plot
plt.show() #show = function to show the plot
```

**BUBBLE PLOT**

```python
# Sample data for the bubble plot
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 25]
bubble_sizes = [50, 100, 80, 150, 120]

# Creating the bubble plot
plt.scatter(x, y, s=bubble_sizes, alpha=0.5)

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bubble Plot Example')

# Displaying the plot
plt.show()
```

```python

# Sample data for cities, population, and average temperature
cities = ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
population = [8622698, 3990456, 2716000, 2312717, 1680992]
average_temperature = [13.9, 18.3, 11.9, 22.0, 24.4]

# Normalize population values to control bubble size
max_population = max(population)
normalized_population = [p / max_population * 1000 for p in population]  # Scaling population to set bubble size

# Create the bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(average_temperature, population, s=normalized_population, alpha=0.7, c='green')

# Adding labels for cities
for i in range(len(cities)):
    plt.text(average_temperature[i], population[i], cities[i], fontsize=10, ha='center', va='center')

# Adding labels and title
plt.xlabel('Average Temperature (Celsius)')
plt.ylabel('Population')
plt.title('City Population vs. Average Temperature Bubble Plot')

# Displaying the plot
plt.show()
```

**STACKED AREA PLOT**

```python
# Sample data for the stacked area plot
x = np.arange(1, 11)
y1 = [2, 3, 4, 3, 5, 6, 7, 8, 6, 5]
y2 = [1, 2, 1, 3, 2, 4, 2, 5, 4, 3]
y3 = [3, 4, 2, 5, 4, 6, 5, 4, 5, 6]

# Creating the stacked area plot
plt.stackplot(x, y1, y2, y3, labels=['A', 'B', 'C'])

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Stacked Area Plot Example')

# Adding legend
plt.legend(loc='upper left')

# Displaying the plot
plt.show()
```

**HISTOGRAM**

```python
# Sample exam scores
exam_scores = [65, 72, 85, 88, 92, 75, 78, 80, 68, 90, 82, 85, 78, 70, 74, 84, 88, 90, 92, 72, 78, 85, 78, 80, 82]

# Creating the histogram
plt.hist(exam_scores, bins=5, edgecolor='black')

# Adding labels and title
plt.xlabel('Exam Scores')
plt.ylabel('Frequency')
plt.title('Exam Scores Distribution')

# Displaying the plot
plt.show()
```

**HEATMAP / CORRELATION MAP**

```python
# Sample correlation matrix
data = {
    'A': [1.0, 0.7, 0.4],
    'B': [0.7, 1.0, 0.9],
    'C': [0.4, 0.9, 1.0]
}

# Convert the data dictionary to a DataFrame
df = pd.DataFrame(data, index=['A', 'B', 'C'])

# Create the heatmap plot
sns.heatmap(df, annot=True, cmap='coolwarm', fmt='.2f')

# Adding title
plt.title('Correlation Heatmap')

# Displaying the plot
plt.show()
```

**PIE PLOT / DONUT PLOT**

```python
# Sample data for market share of smartphone brands
brands = ['Samsung', 'Apple', 'Xiaomi', 'Huawei', 'Others']
market_share = [30, 25, 15, 10, 20]

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(market_share, labels=brands, autopct='%1.1f%%', startangle=140)

# Adding title
plt.title('Market Share of Smartphone Brands')

# Displaying the plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Sample data for market share of smartphone brands
brands = ['Samsung', 'Apple', 'Xiaomi', 'Huawei']
market_share = [25, 20, 15, 10]

# Create the donut chart
plt.figure(figsize=(8, 8))
colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red']
total_share = sum(market_share)
plt.pie(market_share, labels=brands, colors=colors, startangle=90, wedgeprops=dict(width=0.35, edgecolor='w'), autopct=lambda p: f'{p*total_share/100:.0f}')

# Adding a white circle at the center to create the donut effect
plt.gca().add_artist(plt.Circle((0, 0), 0.2, color='white'))

# Adding title
plt.title('Market Share of Smartphone Brands')

# Displaying the plot
plt.axis('equal')  # Equal aspect ratio ensures that the donut is drawn as a circle.
plt.show()
```
